```python
import json
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load data
data = np.array(pd.read_csv("../data/synthetic_100_100_9.csv"))

# Load results 
g = open('synthetic_100_100_9_results.json')
results = json.load(g)
row_partition = results["rowMembership"]
col_partition = results["colMembership"]


# Print results
print("Runtime (sec): {} ".format(results["runtime"]))
print("ARI: ", results["ari"])
print("NMI: ", results["nmi"])
print("Number of inferred clusters: ", results["nClusters"])
```

```python
# Visualization


# Reordering first the rows and then the columns.
reordered_rows = data[np.argsort(row_partition)]
reordered_data = reordered_rows[:, np.argsort(col_partition)]

fig, ax = plt.subplots(1,2, figsize=(8,4))
ax[0].matshow(data, cmap=plt.cm.Greys)
ax[0].title.set_text("Raw data")
ax[1].matshow(reordered_data, cmap=plt.cm.Greys)
ax[1].title.set_text("Reordered data")
plt.savefig("coclustExample")
plt.show();
```
